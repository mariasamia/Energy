```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import LSTM, Bidirectional, Dense 
from keras.layers import Input
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load your dataset
file_path = 'Desktop/data_energy - Copy - Copy - Copy (2).xlsx'
data = pd.read_excel(file_path)

# Preprocessing: Handle missing values
data.ffill(inplace=True)

# Define features and target
X = data[['GDP', 'P.G.', 'CO2']].values
y = data['PES'].values

# Scaling the features using DataFrame (to preserve feature names)
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(data[['GDP', 'P.G.', 'CO2']])  # Fit on the DataFrame

# Scaling the target
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))  # Reshaping y for scaling

# Reshaping the data for LSTM (samples, time steps, features)
X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))


# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, shuffle=False)

# Build the Bidirectional LSTM model
model = Sequential()
model.add(Input(shape=(X_train.shape[1], 1)))  # No need for input_shape in Bidirectional LSTM
model.add(Bidirectional(LSTM(units=50, return_sequences=True)))
model.add(Bidirectional(LSTM(units=50)))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

# Predict the energy consumption on the test set
forecast_scaled = model.predict(X_test)

# Invert scaling to get the actual energy values
forecast_actual = scaler_y.inverse_transform(forecast_scaled)

# Optional: Plot the forecasted values vs actual values for visualization
plt.figure(figsize=(10, 6))
plt.plot(scaler_y.inverse_transform(y_test.reshape(-1, 1)), label='Actual PES', linestyle='-')
plt.plot(forecast_actual, label='Forecasted PES', linestyle='--')
plt.xlabel('Time')
plt.ylabel('PES (TWh)')
plt.title('PES Forecast vs Actual')
plt.legend()
plt.show()

# 1. Forecasting Future Values for 2024 to 2050

# Simulating future values for GDP, P.G., and CO2 from 2024 to 2050
future_years = np.arange(2024, 2051)
future_GDP = np.linspace(X[-1, 0] * 1.02, X[-1, 0] * 1.5, num=27)  # Simulated GDP values
future_PG = np.linspace(X[-1, 1], X[-1, 1] * 0.9, num=27)  # Simulated Population Growth
future_CO2 = np.linspace(X[-1, 2], X[-1, 2] * 1.1, num=27)  # Simulated CO2 Emissions

# Creating a DataFrame for future data
future_data = pd.DataFrame({
    'Year': future_years,
    'GDP': future_GDP,
    'P.G.': future_PG,
    'CO2': future_CO2
})

# Similarly, scale the future data using the same scaler_X
future_X_scaled = scaler_X.transform(future_data[['GDP', 'P.G.', 'CO2']])  # Use the same scaler instance

# Reshape the future data for LSTM input
future_X_scaled = future_X_scaled.reshape((future_X_scaled.shape[0], future_X_scaled.shape[1], 1))

# Predict future PES
future_predictions_scaled = model.predict(future_X_scaled)

# Invert scaling to get the actual future PES values
future_predictions = scaler_y.inverse_transform(future_predictions_scaled)

# 2. Plotting the forecasted future values

# Plotting the future values of CO2, P.G., GDP, and PES
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(future_years, future_GDP, label='Forecasted GDP', color='g')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('Forecasted GDP (2024-2050)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(future_years, future_PG, label='Forecasted Population Growth (P.G.)', color='b')
plt.xlabel('Year')
plt.ylabel('Population Growth')
plt.title('Forecasted Population Growth (2024-2050)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(future_years, future_CO2, label='Forecasted CO2 Emissions', color='r')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions')
plt.title('Forecasted CO2 Emissions (2024-2050)')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(future_years, future_predictions, label='Forecasted PES', color='purple')
plt.xlabel('Year')
plt.ylabel('PES (TWh)')
plt.title('Forecasted PES (2024-2050)')
plt.grid(True)

plt.tight_layout()
plt.show()

# 3. Displaying forecast results in a DataFrame
forecast_df = pd.DataFrame({
    'Year': future_years,
    'Forecasted GDP': future_GDP,
    'Forecasted P.G.': future_PG,
    'Forecasted CO2': future_CO2,
    'Forecasted PES': future_predictions.flatten()  # Flattening to remove extra dimensions
})

# Display the forecasted values for the future years
print(forecast_df)
```
